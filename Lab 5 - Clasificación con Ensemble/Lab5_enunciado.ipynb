3. La matriz de confusión es una matriz cuadrada que compara las clases predichas con las verdaderas, en el contexto de modelos de clasificación con aprendizaje supervisado. Si los valores a predecir se clasifican en $n$ clases, entonces la matriz es de tamaño $n \times n$, tal que en un eje se encuentran las clases reales de las instancias (usualmente el vertical, correspondiente a las filas), y en el otro eje se encuentran las predicciones. Así, la componente $(i,j)$ corresponde a la cantidad de datos que pertenecen a la clase $i$ pero la predicción fue $j$. Es decir, las predicciones correctas del modelo se encuentran en la diagonal, mientras que todo el resto de las casillas corresponde a errores de predicción.

4. En el contexto de la clasificación con aprendizaje supervisado, interesa conocer qué tan acertadas son las predicciones, realizando comparaciones entre los datos correctamente predichos y el total. De allí nacen las siguientes métricas para evaluar la calidad de la predicción de un modelo. Primero definimos la siguiente notación (supondremos que existen dos categorías, positivo y negativo, pero con suficiente cuidado es generalizable a $n$):

    * $TP$: La cantidad de datos que fueron correctamente asignados como positivos (true positives).
    * $TN$: La cantidad de datos que fueron correctamente asignados como negativos (true negatives).
    * $FP$: La cantidad de datos que fueron incorrectamente asignados como positivos (false positives).
    * $FN$: La cantidad de datos que fueron incorrectamente asignados como negativos (false negatives).

Con ello, se definen las siguientes métricas y sus respectivas fórmulas:

* Accuracy: Mide la proporción de los datos correctamente etiquetados sobre el total.

$$
Accuracy = \frac{TP+TN}{TP+TN+FP+FN}
$$

* Precision: De todos los datos catalogados como positivos, mide la proporción de ellos que efectivamente están bien etiquetados.

$$
Precision = \frac{TP}{TP+FP}
$$

* Recall: Mide la proporción de datos correctamente catalogados como positivos sobre todos los datos realmente positivos.

$$
Recall = \frac{TP}{TP+FN}
$$

* F1 score: Corresponde a la media armónica entre las métricas precision y recall. Es una métrica más balanceada cuando no se prioriza la minimización de un error a costa del otro.

$$
F_1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall} = \frac{2TP}{2TP+FP+FN}
$$

5. Las métricas recomendadas a optimizar para cada caso son:

    * Mantenimiento predictivo de fallas de maquinaria pesada en la industria minera: Depende de qué es más costoso, una detención por falla o una serie de detenciones preventivas a cambio de prevenir fallas de maquinaria. Suponiendo que es lo primero entonces conviene optimizar Recall, pues lo ideal es identificar todos las fallas posibles, aún si a veces se predicen algunos falsos positivos. Dicho de otro modo, el objetivo principal es minimizar los falsos negativos.
    * Detección de enfermedades altamente contagiosas: Recall. Al igual que en el caso anterior, lo que se busca es minimizar los falsos negativos, aún si eso genera un leve incremento en la cantidad de falsos positivos.
    * Aprobación de créditos de alto riesgo: Precision. No es tan grave dejar sin un crédito de alto riesgo a alguien calificado, pero el banco no se puede permitir que una persona riesgosa reciba uno (falso positivo). En ese sentido, la precisión es una métrica un poco más conservadora. 
    * Detección de crímenes: F1 score. Un gobierno o la policía local de un país no pueden permitirse encarcelar a alguien por un crimen que no cometió, pero también es muy necesario detectar la mayor cantidad de criminales, de modo que el F1 score provee ese balance.

6. La calibración de modelos es el proceso mediante el cual verificamos si las probabilidades que entrega un modelo de clasificación realmente reflejan la realidad. En otras palabras, queremos saber si cuando el modelo dice, por ejemplo, que hay un 80% de probabilidad de que algo ocurra, eso significa que en efecto, en promedio, eso ocurre el 80% de las veces.
    
    Esto es especialmente importante en modelos que entregan probabilidades como salida, porque no basta con que el modelo acierte o se equivoque en su clasificación: también importa qué tan confiado está y si esa confianza está justificada.

    Un modelo está bien calibrado si, por ejemplo, de todos los casos en los que asigna un 70% de probabilidad a una clase, aproximadamente el 70% de esos casos realmente pertenecen a esa clase. Si asigna probabilidades muy altas pero se equivoca frecuentemente, o al revés, sugiere estar poco seguro pero acierta mucho, entonces el modelo está mal calibrado.

````python
numerical = data.select_dtypes(include=['int']).columns.tolist()
categorical = data.select_dtypes(include=['object']).columns.tolist()

numerical.remove('label')
````

````python
# Define imputadores para valores faltantes
cat_imputer = SimpleImputer(strategy='most_frequent')  
num_imputer = SimpleImputer(strategy='mean')          

# Define el ColumnTransformer con imputación y transformación
col_transformer = ColumnTransformer(
    transformers=[
        ('cat', Pipeline([
            ('imputer', cat_imputer),  # Imputar valores faltantes
            ('encoder', OneHotEncoder(handle_unknown='ignore'))  # Codificar categóricas
        ]), categorical),
        ('num', Pipeline([
            ('imputer', num_imputer),  # Imputar valores faltantes
            ('scaler', StandardScaler())  # Escalar numéricas
        ]), numerical)])
````

### 1.2 Entrenamiento [3 puntos]

Ahora, vamos a entrenar los pipelines generados en los pasos anteriores. Para esto, debe realizar las siguientes tareas:

**Respuesta:**

````python
x = data.drop(columns=['label'])
y = data['label']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)


pipeline_xgboost = Pipeline(
    steps=[
        ('Preprocessor', col_transformer),
        ('Classifier', XGBClassifier(random_state=42))
    ]
)

pipeline_lightgbm = Pipeline(
    steps=[
        ('Preprocessor', col_transformer),
        ('Classifier', LGBMClassifier(random_state=42))
    ]
)


pipeline_xgboost.fit(x_train, y_train)
pipeline_lightgbm.fit(x_train, y_train)
````

La elección del clasificador se justificar en base a la siguiente [guía](https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html) y se deben comentar los resultados obtenidos en la clasificación.

**Tareas:** [1 punto por tarea]

1. En un nuevo dataframe, aplique las etiquetas descritas anteriormente en cada uno de los valores señalados en esta sección y guárdelos en la variable `label`.
2. Cuente cuántos por clase quedan.
3. Entrene el nuevo pipeline y ejecute una evaluación de este.  
4. Comente los resultados obtenidos.

**Respuesta:**

````python
df = pd.read_csv('stats_players.csv')

ataque = ['ST', 'CF']
central_ataque = ['RW', 'CAM', 'LW']
central = ['RM', 'CM', 'LM']
central_defensa = ['RWB', 'CDM', 'LWB']
defensa = ['RB', 'CB', 'LB']
arquero = ['GK']

def label_position(position):
    if position in ataque:
        return 'Ataque'
    elif position in central_ataque:
        return 'Central Ataque'
    elif position in central:
        return 'Central'
    elif position in central_defensa:
        return 'Central Defensa'
    elif position in defensa:
        return 'Defensa'
    elif position in arquero:
        return 'Arquero'
     

df['label'] = df['National_Position'].apply(label_position)
df = df[df['label'].notnull()]
df['label'].value_counts()  

numerical1 = df.select_dtypes(include=['int']).columns.tolist()
categorical1 = df.select_dtypes(include=['object']).columns.tolist()
categorical1.remove('label')
````

````python
# Define imputadores para valores faltantes
cat_imputer = SimpleImputer(strategy='most_frequent')  
num_imputer = SimpleImputer(strategy='mean')          

# Define el ColumnTransformer con imputación y transformación
col_transformer = ColumnTransformer(
    transformers=[
        ('cat', Pipeline([
            ('imputer', cat_imputer),  # Imputar valores faltantes
            ('encoder', OneHotEncoder(handle_unknown='ignore'))  # Codificar categóricas
        ]), categorical1),
        ('num', Pipeline([
            ('imputer', num_imputer),  # Imputar valores faltantes
            ('scaler', StandardScaler())  # Escalar numéricas
        ]), numerical1)])
````

````python
label_encoder = LabelEncoder()
df['label_encoded'] = label_encoder.fit_transform(df['label'])


x1 = df.drop(columns=['label'])
y1 = df['label_encoded']
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size=0.2, random_state=42, stratify=y1)


pipeline_xgboost1 = Pipeline(
    steps=[
        ('Preprocessor', col_transformer),
        ('Classifier', XGBClassifier(random_state=42, verbose=0))
    ]
)

pipeline_lightgbm1 = Pipeline(
    steps=[
        ('Preprocessor', col_transformer),
        ('Classifier', LGBMClassifier(random_state=42, verbose=0))
    ]
)

pipeline_xgboost1.fit(x_train1, y_train1)
pipeline_lightgbm1.fit(x_train1, y_train1)


y_pred_xgboost1 = pipeline_xgboost1.predict(x_test1)
y_pred_lightgbm1 = pipeline_lightgbm1.predict(x_test1)

classification_report_xgboost1 = classification_report(y_test1, y_pred_xgboost1)
classification_report_lightgbm1 = classification_report(y_test1, y_pred_lightgbm1)
accuracy_xgboost1 = accuracy_score(y_test1,y_pred_xgboost1)
accuracy_lightgbm1 = accuracy_score(y_test1,y_pred_lightgbm1)
````

````python
print('XGBoost Classification Report')
print(classification_report_xgboost1)
print('XGBoost Accuracy:', accuracy_xgboost1)

print('LightGBM Classification Report')
print(classification_report_lightgbm1)
print('LightGBM Accuracy:', accuracy_lightgbm1)
````

````python
from xgboost import XGBClassifier
from sklearn.feature_selection import RFECV
from sklearn.metrics import classification_report, accuracy_score

# Entrenar un modelo XGBoost con todas las características disponibles
xgb_model_full = XGBClassifier(random_state=42)
xgb_model_full.fit(x_train, y_train)

# Evaluar y clasificar las características según su importancia
feature_importances = xgb_model_full.feature_importances_
sorted_features = sorted(zip(x_train.columns, feature_importances), key=lambda x: x[1], reverse=True)

print("Características ordenadas por importancia:")
for feature, importance in sorted_features:
    print(f"{feature}: {importance:.4f}")

# Búsqueda recursiva de eliminación de características (RFE)
rfe_selector = RFECV(estimator=xgb_model_full, step=1, cv=5, scoring='accuracy')
rfe_selector.fit(x_train, y_train)

# Seleccionar las características más importantes
selected_features = x_train.columns[rfe_selector.support_]
print("\nCaracterísticas seleccionadas:")
print(selected_features)

# Entrenar un nuevo modelo con el conjunto reducido de características
x_train_reduced = x_train[selected_features]
x_test_reduced = x_test[selected_features]

xgb_model_reduced = XGBClassifier(random_state=42)
xgb_model_reduced.fit(x_train_reduced, y_train)

# Evaluar el modelo reducido
y_pred_reduced = xgb_model_reduced.predict(x_test_reduced)
print("\nReporte de clasificación para el modelo reducido:")
print(classification_report(y_test, y_pred_reduced))

# Comparar rendimiento entre el modelo original y el modelo reducido
y_pred_full = xgb_model_full.predict(x_test)
print("\nReporte de clasificación para el modelo completo:")
print(classification_report(y_test, y_pred_full))

accuracy_full = accuracy_score(y_test, y_pred_full)
accuracy_reduced = accuracy_score(y_test, y_pred_reduced)

print(f"\nPrecisión del modelo completo: {accuracy_full:.4f}")
print(f"Precisión del modelo reducido: {accuracy_reduced:.4f}")
````

Para lograr modelos más modulares, se recomienda realizar una calibración del modelo entrenado anteriormente, con el objetivo de obtener salidas que reflejen mayor modularidad.

1. Se solicita que utilice un método de calibración que asegure que las probabilidades generadas incrementen de manera monótona. Una métrica ampliamente utilizada para evaluar la precisión de la calibración de un modelo es el Brier Score. Calcule el Brier Score para el modelo tanto antes como después de la calibración. Esto le permitirá realizar una comparación cuantitativa y determinar si la calibración ha mejorado el rendimiento del modelo. Para más información sobre el Brier Score, puede consultar el siguiente enlace: [Scikit-Learn - Brier Score Loss](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.brier_score_loss.html). 2. Tras la calibración, examine y comente los resultados obtenidos. A su análisis añada una comparación visual de las ideales versus las salidas del modelo original (sin calibrar) y del modelo calibrado. [3 puntos]

**Respuesta**

````python
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import brier_score_loss
import numpy as np

import matplotlib.pyplot as plt

# Entrenar el modelo XGBoost con los datos de entrenamiento
xgb_model = XGBClassifier(random_state=42)
xgb_model.fit(x_train, y_train)

# Calcular el Brier Score antes de la calibración
y_prob_before = xgb_model.predict_proba(x_test)[:, 1]
brier_score_before = brier_score_loss(y_test, y_prob_before)

# Calibrar el modelo utilizando CalibratedClassifierCV con método isotónico
calibrated_model = CalibratedClassifierCV(xgb_model, method='isotonic', cv='prefit')
calibrated_model.fit(x_test, y_test)

# Calcular el Brier Score después de la calibración
y_prob_after = calibrated_model.predict_proba(x_test)[:, 1]
brier_score_after = brier_score_loss(y_test, y_prob_after)

# Comparación visual de las probabilidades ideales vs. las salidas del modelo
plt.figure(figsize=(10, 6))
plt.plot([0, 1], [0, 1], 'k--', label='Ideal')
plt.hist(y_prob_before, bins=50, alpha=0.5, label='Antes de calibración', density=True)
plt.hist(y_prob_after, bins=50, alpha=0.5, label='Después de calibración', density=True)
plt.xlabel('Probabilidad predicha')
plt.ylabel('Densidad')
plt.title('Comparación de calibración del modelo')
plt.legend()
plt.show()

# Imprimir los resultados del Brier Score
print(f"Brier Score antes de la calibración: {brier_score_before:.4f}")
print(f"Brier Score después de la calibración: {brier_score_after:.4f}")
````
