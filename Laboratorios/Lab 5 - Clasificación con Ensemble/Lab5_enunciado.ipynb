3. La matriz de confusi√≥n es una matriz cuadrada que compara las clases predichas con las verdaderas, en el contexto de modelos de clasificaci√≥n con aprendizaje supervisado. Si los valores a predecir se clasifican en $n$ clases, entonces la matriz es de tama√±o $n \times n$, tal que en un eje se encuentran las clases reales de las instancias (usualmente el vertical, correspondiente a las filas), y en el otro eje se encuentran las predicciones. As√≠, la componente $(i,j)$ corresponde a la cantidad de datos que pertenecen a la clase $i$ pero la predicci√≥n fue $j$. Es decir, las predicciones correctas del modelo se encuentran en la diagonal, mientras que todo el resto de las casillas corresponde a errores de predicci√≥n.

4. En el contexto de la clasificaci√≥n con aprendizaje supervisado, interesa conocer qu√© tan acertadas son las predicciones, realizando comparaciones entre los datos correctamente predichos y el total. De all√≠ nacen las siguientes m√©tricas para evaluar la calidad de la predicci√≥n de un modelo. Primero definimos la siguiente notaci√≥n (supondremos que existen dos categor√≠as, positivo y negativo, pero con suficiente cuidado es generalizable a $n$):

    * $TP$: La cantidad de datos que fueron correctamente asignados como positivos (true positives).
    * $TN$: La cantidad de datos que fueron correctamente asignados como negativos (true negatives).
    * $FP$: La cantidad de datos que fueron incorrectamente asignados como positivos (false positives).
    * $FN$: La cantidad de datos que fueron incorrectamente asignados como negativos (false negatives).

Con ello, se definen las siguientes m√©tricas y sus respectivas f√≥rmulas:

* Accuracy: Mide la proporci√≥n de los datos correctamente etiquetados sobre el total.

$$
Accuracy = \frac{TP+TN}{TP+TN+FP+FN}
$$

* Precision: De todos los datos catalogados como positivos, mide la proporci√≥n de ellos que efectivamente est√°n bien etiquetados.

$$
Precision = \frac{TP}{TP+FP}
$$

* Recall: Mide la proporci√≥n de datos correctamente catalogados como positivos sobre todos los datos realmente positivos.

$$
Recall = \frac{TP}{TP+FN}
$$

* F1 score: Corresponde a la media arm√≥nica entre las m√©tricas precision y recall. Es una m√©trica m√°s balanceada cuando no se prioriza la minimizaci√≥n de un error a costa del otro.

$$
F_1 = 2 \cdot \frac{Precision \cdot Recall}{Precision + Recall} = \frac{2TP}{2TP+FP+FN}
$$

5. Las m√©tricas recomendadas a optimizar para cada caso son:

Suponiendo que es lo primero entonces conviene optimizar Recall, pues lo ideal es identificar todos las fallas posibles, a√∫n si a veces se predicen algunos falsos positivos. Dicho de otro modo, el objetivo principal es minimizar los falsos negativos.
    * Detecci√≥n de enfermedades altamente contagiosas: Recall. Al igual que en el caso anterior, lo que se busca es minimizar los falsos negativos, a√∫n si eso genera un leve incremento en la cantidad de falsos positivos.
    * Aprobaci√≥n de cr√©ditos de alto riesgo: Precision. No es tan grave dejar sin un cr√©dito de alto riesgo a alguien calificado, pero el banco no se puede permitir que una persona riesgosa reciba uno (falso positivo). En ese sentido, la precisi√≥n es una m√©trica un poco m√°s conservadora. 
    * Detecci√≥n de cr√≠menes: F1 score. Un gobierno o la polic√≠a local de un pa√≠s no pueden permitirse encarcelar a alguien por un crimen que no cometi√≥, pero tambi√©n es muy necesario detectar la mayor cantidad de criminales, de modo que el F1 score provee ese balance.

6. La calibraci√≥n de modelos es el proceso mediante el cual verificamos si las probabilidades que entrega un modelo de clasificaci√≥n realmente reflejan la realidad. En otras palabras, queremos saber si cuando el modelo dice, por ejemplo, que hay un 80% de probabilidad de que algo ocurra, eso significa que en efecto, en promedio, eso ocurre el 80% de las veces.
    
    Esto es especialmente importante en modelos que entregan probabilidades como salida, porque no basta con que el modelo acierte o se equivoque en su clasificaci√≥n: tambi√©n importa qu√© tan confiado est√° y si esa confianza est√° justificada.

    Un modelo est√° bien calibrado si, por ejemplo, de todos los casos en los que asigna un 70% de probabilidad a una clase, aproximadamente el 70% de esos casos realmente pertenecen a esa clase. Si asigna probabilidades muy altas pero se equivoca frecuentemente, o al rev√©s, sugiere estar poco seguro pero acierta mucho, entonces el modelo est√° mal calibrado.

````python
data['label'] = data['National_Position'].notnull().astype(int)
data['label'].value_counts()
````

Del total de jugadores, hay 1075 que son seleccionados nacionales, lo que corresponde a aproximadamente el 6.1% de los datos. Esto indica que si queremos predecir los jugadores que son seleccionados nacionales, debemos tener un alto nivel en las m√©tricas para que el modelo se pueda considerar relevante.

````python
numerical = data.select_dtypes(include=['int']).columns.tolist()
categorical = data.select_dtypes(include=['object']).columns.tolist()

numerical.remove('label')
````

````python
# Define imputadores para valores faltantes
cat_imputer = SimpleImputer(strategy='most_frequent')  
num_imputer = SimpleImputer(strategy='mean')          

# Define el ColumnTransformer con imputaci√≥n y transformaci√≥n
col_transformer = ColumnTransformer(
    transformers=[
        ('cat', Pipeline([
            ('imputer', cat_imputer),  # Imputar valores faltantes
            ('encoder', OneHotEncoder(handle_unknown='ignore'))  # Codificar categ√≥ricas
        ]), categorical),
        ('num', Pipeline([
            ('imputer', num_imputer),  # Imputar valores faltantes
            ('scaler', StandardScaler())  # Escalar num√©ricas
        ]), numerical)])
````

### 1.2 Entrenamiento [3 puntos]

Ahora, vamos a entrenar los pipelines generados en los pasos anteriores. Para esto, debe realizar las siguientes tareas:

1. Separe los datos de entrenamiento en un conjunto de entrenamiento y de prueba  (la proporci√≥n queda a su juicio). En este paso, seleccione los ejemplos de forma aleatoria e intente mantener la distribuci√≥n original de labels de cada clase en los conjuntos de prueba/entrenamiento. 3. Entrene los pipelines. [1 punto]



**Respuesta:**

````python
x = data.drop(columns=['label'])
y = data['label']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)


pipeline_xgboost = Pipeline(
    steps=[
        ('Preprocessor', col_transformer),
        ('Classifier', XGBClassifier(random_state=42))
    ]
)

pipeline_lightgbm = Pipeline(
    steps=[
        ('Preprocessor', col_transformer),
        ('Classifier', LGBMClassifier(random_state=42))
    ]
)


pipeline_xgboost.fit(x_train, y_train)
pipeline_lightgbm.fit(x_train, y_train)
````

Y adem√°s, que hay bastantes jugadores nuevos que no tienen muy claro en que posici√≥n verdaderamente brillar√≠an, por lo que actualmente puede que actualmente est√©n jugando en posiciones sub-optimas.

Viendo que los resultados del primer an√°lisis no son tan esperanzadores, el corporeo los comanda a cambiar su tarea: ahora, les solicita que construyan un clasificador enfocado en predecir la mejor posici√≥n de los jugadores en la cancha seg√∫n sus caracter√≠sticas.

Para lograr esto, primero, les pide que etiqueten de la siguiente manera los valores que aparecen en el atributo `Club_Position`, pidiendo que agrupen los valores en los siguientes grupos:

**Nota**:  Renac√≠n les recalca que **no deben utilizar los valores ```Sub``` y ```Res``` de esta columna**.

```python
ataque = ['ST', 'CF']

```

La elecci√≥n del clasificador se justificar en base a la siguiente [gu√≠a](https://scikit-learn.org/stable/tutorial/machine_learning_map/index.html) y se deben comentar los resultados obtenidos en la clasificaci√≥n.

**Tareas:** [1 punto por tarea]

1. En un nuevo dataframe, aplique las etiquetas descritas anteriormente en cada uno de los valores se√±alados en esta secci√≥n y gu√°rdelos en la variable `label`.
2. Cuente cu√°ntos por clase quedan.
3. Entrene el nuevo pipeline y ejecute una evaluaci√≥n de este.  
4. Comente los resultados obtenidos.

**Respuesta:**

````python
df = pd.read_csv('stats_players.csv')

ataque = ['ST', 'CF']
central_ataque = ['RW', 'CAM', 'LW']
central = ['RM', 'CM', 'LM']
central_defensa = ['RWB', 'CDM', 'LWB']
defensa = ['RB', 'CB', 'LB']
arquero = ['GK']

def label_position(position):
    if position in ataque:
        return 'Ataque'
    elif position in central_ataque:
        return 'Central Ataque'
    elif position in central:
        return 'Central'
    elif position in central_defensa:
        return 'Central Defensa'
    elif position in defensa:
        return 'Defensa'
    elif position in arquero:
        return 'Arquero'
     

df['label'] = df['National_Position'].apply(label_position)
df = df[df['label'].notnull()]
df['label'].value_counts()  

numerical1 = df.select_dtypes(include=['int']).columns.tolist()
categorical1 = df.select_dtypes(include=['object']).columns.tolist()
categorical1.remove('label')
````

````python
# Define imputadores para valores faltantes
cat_imputer = SimpleImputer(strategy='most_frequent')  
num_imputer = SimpleImputer(strategy='mean')          

# Define el ColumnTransformer con imputaci√≥n y transformaci√≥n
col_transformer = ColumnTransformer(
    transformers=[
        ('cat', Pipeline([
            ('imputer', cat_imputer),  # Imputar valores faltantes
            ('encoder', OneHotEncoder(handle_unknown='ignore'))  # Codificar categ√≥ricas
        ]), categorical1),
        ('num', Pipeline([
            ('imputer', num_imputer),  # Imputar valores faltantes
            ('scaler', StandardScaler())  # Escalar num√©ricas
        ]), numerical1)])
````

````python
label_encoder = LabelEncoder()
df['label_encoded'] = label_encoder.fit_transform(df['label'])


x1 = df.drop(columns=['label'])
y1 = df['label_encoded']
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y1, test_size=0.2, random_state=42, stratify=y1)


pipeline_xgboost1 = Pipeline(
    steps=[
        ('Preprocessor', col_transformer),
        ('Classifier', XGBClassifier(random_state=42, verbose=0))
    ]
)

pipeline_lightgbm1 = Pipeline(
    steps=[
        ('Preprocessor', col_transformer),
        ('Classifier', LGBMClassifier(random_state=42, verbose=0))
    ]
)

pipeline_xgboost1.fit(x_train1, y_train1)
pipeline_lightgbm1.fit(x_train1, y_train1)


y_pred_xgboost1 = pipeline_xgboost1.predict(x_test1)
y_pred_lightgbm1 = pipeline_lightgbm1.predict(x_test1)

classification_report_xgboost1 = classification_report(y_test1, y_pred_xgboost1)
classification_report_lightgbm1 = classification_report(y_test1, y_pred_lightgbm1)
accuracy_xgboost1 = accuracy_score(y_test1,y_pred_xgboost1)
accuracy_lightgbm1 = accuracy_score(y_test1,y_pred_lightgbm1)
````

````python
print('XGBoost Classification Report')
print(classification_report_xgboost1)
print('XGBoost Accuracy:', accuracy_xgboost1)

print('LightGBM Classification Report')
print(classification_report_lightgbm1)
print('LightGBM Accuracy:', accuracy_lightgbm1)
````

````python
from xgboost import XGBClassifier
from sklearn.feature_selection import RFECV
from sklearn.metrics import classification_report, accuracy_score

# Entrenar un modelo XGBoost con todas las caracter√≠sticas disponibles
xgb_model_full = XGBClassifier(random_state=42)
xgb_model_full.fit(x_train, y_train)

# Evaluar y clasificar las caracter√≠sticas seg√∫n su importancia
feature_importances = xgb_model_full.feature_importances_
sorted_features = sorted(zip(x_train.columns, feature_importances), key=lambda x: x[1], reverse=True)

print("Caracter√≠sticas ordenadas por importancia:")
for feature, importance in sorted_features:
    print(f"{feature}: {importance:.4f}")

# B√∫squeda recursiva de eliminaci√≥n de caracter√≠sticas (RFE)
rfe_selector = RFECV(estimator=xgb_model_full, step=1, cv=5, scoring='accuracy')
rfe_selector.fit(x_train, y_train)

# Seleccionar las caracter√≠sticas m√°s importantes
selected_features = x_train.columns[rfe_selector.support_]
print("\nCaracter√≠sticas seleccionadas:")
print(selected_features)

# Entrenar un nuevo modelo con el conjunto reducido de caracter√≠sticas
x_train_reduced = x_train[selected_features]
x_test_reduced = x_test[selected_features]

xgb_model_reduced = XGBClassifier(random_state=42)
xgb_model_reduced.fit(x_train_reduced, y_train)

# Evaluar el modelo reducido
y_pred_reduced = xgb_model_reduced.predict(x_test_reduced)
print("\nReporte de clasificaci√≥n para el modelo reducido:")
print(classification_report(y_test, y_pred_reduced))

# Comparar rendimiento entre el modelo original y el modelo reducido
y_pred_full = xgb_model_full.predict(x_test)
print("\nReporte de clasificaci√≥n para el modelo completo:")
print(classification_report(y_test, y_pred_full))

accuracy_full = accuracy_score(y_test, y_pred_full)
accuracy_reduced = accuracy_score(y_test, y_pred_reduced)

print(f"\nPrecisi√≥n del modelo completo: {accuracy_full:.4f}")
print(f"Precisi√≥n del modelo reducido: {accuracy_reduced:.4f}")
````

### 3.3 Calibraci√≥n Probabilistica [6 puntos]

Para lograr modelos m√°s modulares, se recomienda realizar una calibraci√≥n del modelo entrenado anteriormente, con el objetivo de obtener salidas que reflejen mayor modularidad.

1. Se solicita que utilice un m√©todo de calibraci√≥n que asegure que las probabilidades generadas incrementen de manera mon√≥tona. Una m√©trica ampliamente utilizada para evaluar la precisi√≥n de la calibraci√≥n de un modelo es el Brier Score. Calcule el Brier Score para el modelo tanto antes como despu√©s de la calibraci√≥n. Esto le permitir√° realizar una comparaci√≥n cuantitativa y determinar si la calibraci√≥n ha mejorado el rendimiento del modelo. Para m√°s informaci√≥n sobre el Brier Score, puede consultar el siguiente enlace: [Scikit-Learn - Brier Score Loss](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.brier_score_loss.html). [3 puntos]

2. Tras la calibraci√≥n, examine y comente los resultados obtenidos. A su an√°lisis a√±ada una comparaci√≥n visual de las ideales versus las salidas del modelo original (sin calibrar) y del modelo calibrado. [3 puntos]

**Respuesta**

````python
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import brier_score_loss
import numpy as np

import matplotlib.pyplot as plt

# Entrenar el modelo XGBoost con los datos de entrenamiento
xgb_model = XGBClassifier(random_state=42)
xgb_model.fit(x_train, y_train)

# Calcular el Brier Score antes de la calibraci√≥n
y_prob_before = xgb_model.predict_proba(x_test)[:, 1]
brier_score_before = brier_score_loss(y_test, y_prob_before)

# Calibrar el modelo utilizando CalibratedClassifierCV con m√©todo isot√≥nico
calibrated_model = CalibratedClassifierCV(xgb_model, method='isotonic', cv='prefit')
calibrated_model.fit(x_test, y_test)

# Calcular el Brier Score despu√©s de la calibraci√≥n
y_prob_after = calibrated_model.predict_proba(x_test)[:, 1]
brier_score_after = brier_score_loss(y_test, y_prob_after)

# Comparaci√≥n visual de las probabilidades ideales vs. las salidas del modelo
plt.figure(figsize=(10, 6))
plt.plot([0, 1], [0, 1], 'k--', label='Ideal')
plt.hist(y_prob_before, bins=50, alpha=0.5, label='Antes de calibraci√≥n', density=True)
plt.hist(y_prob_after, bins=50, alpha=0.5, label='Despu√©s de calibraci√≥n', density=True)
plt.xlabel('Probabilidad predicha')
plt.ylabel('Densidad')
plt.title('Comparaci√≥n de calibraci√≥n del modelo')
plt.legend()
plt.show()

# Imprimir los resultados del Brier Score
print(f"Brier Score antes de la calibraci√≥n: {brier_score_before:.4f}")
print(f"Brier Score despu√©s de la calibraci√≥n: {brier_score_after:.4f}")
````
